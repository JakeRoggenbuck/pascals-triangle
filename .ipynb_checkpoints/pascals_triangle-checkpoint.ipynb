```python
from math import factorial, log, floor, ceil
import math
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_theme()


def pascal(rule=lambda x: x, length: int = 10, show: bool = False):
    full = []
    for i in range(length):
        for j in range(length - i + 1):
            if show:
                print(end=" ")

        row = []
        for j in range(i + 1):
            # nCr = n!/((n-r)!*r!)
            nCr = factorial(i) // (factorial(j) * factorial(i - j))
            mod = rule(nCr)
            row.append(mod)
            if show:
                print(mod, end=" ")
        if show:
            print()

        full.append(row)
    return full
```

```python
def try_one():
    return pascal(rule=lambda x: round(log(x)), length=10, show=True)
```

```python
try_one()
```

```python
def try_two():
    length = 200
    a = pascal(rule=lambda x: round(log(x)), length=length, show=False)
    new = []
    for b in a:
        new.append(([0] * (length - len(b))) + b)

    sns.heatmap(new)
    plt.show()
    return new
```

```python
def try_seventeen():
    length = 200
    rule_a = lambda x: 1 if x % 6 else 0
    rule_b = lambda x: 1 if x % 11 else 0

    a = pascal(rule=rule_a, length=length)
    b = pascal(rule=rule_b, length=length)

    c = []
    for x, y in zip(a, b):
        h = []
        for j, k in zip(x, y):
            h.append(j - k)
        c.append(h)
            
    # print(c)
    
    new = []
    for p in c:
        before = [0] * (floor(length - len(p) / 2))
        after = [0] * (ceil(length - len(p) / 2))
        cur = before + p + after
        new.append(cur)

    sns.heatmap(new)
    plt.show()
    return new
```

```python
def try_nineteen():
    length = 200
    rule_a = lambda x: 1 if x % 6 else 0
    rule_b = lambda x: 1 if x % 11 else 0

    a = pascal(rule=rule_a, length=length)
    b = pascal(rule=rule_b, length=length)

    c = []
    for x, y in zip(a, b):
        h = []
        for j, k in zip(x, y):
            h.append(abs(j - k))
        c.append(h)
            
    # print(c)
    
    new = []
    for p in c:
        before = [0] * (floor(length - len(p) / 2))
        after = [0] * (ceil(length - len(p) / 2))
        cur = before + p + after
        new.append(cur)

    sns.heatmap(new)
    plt.show()
    return new
```

```python
from PIL import Image
from IPython.display import display

def to_image(array):
    w = h = len(array) * 2
    data = np.zeros((h, w, 3), dtype=np.uint8)
    
    for x, row in enumerate(array):
        for y, column in enumerate(array[x]):
            data[x, y] = (210, 200, 200) if column else 0

    img = Image.fromarray(data, "RGB")
    display(img)
```

```python
def try_twenty_two(f, g):
    length = 200
    rule_a = lambda x: 1 if x % f else 0
    rule_b = lambda x: 1 if x % g else 0

    a = pascal(rule=rule_a, length=length)
    b = pascal(rule=rule_b, length=length)

    c = []
    for x, y in zip(a, b):
        h = []
        for j, k in zip(x, y):
            h.append(abs(1 - math.sin(1) / max(k, 0.001)))
        c.append(h)
            
    # print(c)
    
    new = []
    for p in c:
        before = [0] * (floor(length - len(p) / 2))
        after = [0] * (ceil(length - len(p) / 2))
        cur = before + p + after
        new.append(cur)

    sns.heatmap(new)
    plt.show()
    return new
```

```python
out = try_twenty_two(100, 10)
```
